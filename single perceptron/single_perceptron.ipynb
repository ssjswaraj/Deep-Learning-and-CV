```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/deep learning/single perceptron/placement.csv")
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(data=df, x='cgpa', y='resume_score', hue='placed')
```

```python
df.head()
```

```python
x=df.iloc[:,:2]
y=df.iloc[:,-1:]
```

```python
from sklearn.linear_model import Perceptron
```

```python
p=Perceptron()
```

```python
p.fit(x,y)
```

```python
p.coef_, p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.values, y.values.flatten(), clf=p, legend=2)
```

```python
p1=Perceptron(penalty ='elasticnet')
```

```python
p1.fit(x,y)
```

```python
p1.coef_ , p1.intercept_
```

```python
plot_decision_regions(x.values,y.values.flatten(),clf=p1)
```
